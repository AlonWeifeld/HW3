```python
   
#B  
def longest_words(file):
    try:
        data = open(file)
        last_words = []
        for line in data:
            line = line.strip().split()
            for word in line:
                word = word.split(".")
                for w_without_p in word:
                    w_without_p =  w_without_p.split("-")
                    for final_w in  w_without_p:
                        last_words.append(final_w)
        
        sort_words = sorted(last_words, key = len, reverse= True)
        return sort_words[:5] 
    except IOError:
        print("File not found")               
                     
longest_words("D:\Alon/ex3_text.txt")    
```
